```python
import statistics as st
import itertools as it
```

```python
pop = [5,1,4,10,7]
```

```python
popmean = st.mean(pop)
print("Thr population mean is :", popmean)
```

The population mean ̅Y is 5.4.

```python
popvar = st.pvariance(pop)
print("The population variance is :", popvar)
```

The population sample variance S^2 = 9.04

```python
comb3 = list(it.combinations(pop,3))
print(comb3)
```

hww16.png

```python
print("Combination | Sample Mean | Sample Variance")
print("-------------------------------------------")
for comb in comb3:
    sampmean = sum(comb) / len(comb)
    sampvar = st.variance(comb)
    print(f"{comb} | {sampmean:.2f} | {sampvar:.2f}")
```

# Population mean was 5.4 and the poulation variance was 9.04. So no, none of the sample means or element variances are equal to thir corresponding population parameters.

```python

sampmeanlist = [3.33, 5.33, 4.33, 6.33, 5.33, 7.33, 5.00, 4.00, 6.00, 7.00]
sampvarlist = [4.33, 20.33, 9.33, 10.33, 2.33, 6.33, 21.00, 9.00, 21.00, 9.00]
avgmean = st.mean(sampmeanlist)
avgvar = st.mean(sampvarlist)

print("The average of the sample means is :", avgmean)
print("The average of the element variances is :", avgvar)
```

# The average of the sample means is 5.398, which is the same as the population mean 5.4. The average of the element variances is 11.298 which is not the same as the population variance of 9.04.
