###  Q1 A sample size of 1547 is required so that the total length of a 95% CI for proprtion of EVERPREG is less than 0.05 (total required sample size) is 1547.

```python
p = 0.6285  # Proportion of EVERPREG = 1 from HOMEWORK 3
Z = 1.96
CI_length = 0.05
DEFF_everpreg = 1.0779  # DEFF for EVERPREG

# Required sample size for SRS
n_srs = (Z**2 * p * (1 - p)) / (CI_length / 2) ** 2

# Adjust for design effect
n_required = n_srs * DEFF_everpreg

# Given proportionate allocation in Homework 3
prop_allocation = np.array([9.97419355, 11.10967742, 9.74193548, 5.52258065, 3.6516129])
prop_allocation_scaled = (prop_allocation / sum(prop_allocation)) * n_required

# Equal allocation (divide total sample size equally among 5 strata)
equal_allocation = np.full(5, n_required / 5)

# Output results
round(n_required), np.round(equal_allocation), np.round(prop_allocation_scaled)

```
